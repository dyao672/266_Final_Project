```python
import os
import pandas as pd

# Directory path
folder_path = '/Users/TennisPhysics/Downloads/archive (1)'

# List all files in the folder
files = os.listdir(folder_path)

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each file in the folder
for file in files:
    # Check if the file is a CSV file
    if file.endswith('.csv'):
        # Create the full file path
        file_path = os.path.join(folder_path, file)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # df.drop('Genre', axis=1, inplace=True)
        
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames into one
concatenated_df = pd.concat(dfs, ignore_index=True)

# # Save the concatenated DataFrame to a new CSV file
# concatenated_df.to_csv('/Users/TennisPhysics/Downloads/movies_metadata.csv', index=False)

# # Optional: Print out the first few rows of the concatenated DataFrame
# print("Movies CSV created with all data:")
concatenated_df.head()
```

```python
movie_data = pd.read_csv('/Users/TennisPhysics/Downloads/movie_data.csv')
movie_data.head(5)
```

```python
# Function to safely create the 'movie_name_year' column
def create_movie_name_year(row):
    # Check if both movie_name and year are valid
    if pd.notnull(row['movie_name']) and pd.notnull(row['year']):
        try:
            # Try converting the year to an integer
            year = int(row['year'])
            return f"{row['movie_name']} ({year})"
        except ValueError:
            # If conversion fails, return None or a default string
            return f"{row['movie_name']} (Invalid Year)"
    return None

# Step 2: Apply the function to create the new column
concatenated_df['movie_name_year'] = concatenated_df.apply(create_movie_name_year, axis=1)

# Display the DataFrame
concatenated_df.head(5)
```

```python
movies = df_no_duplicates.copy()
```

```python
# Split the 'genre' column into up to 3 columns
df = movies

# Split the 'genre' column into up to 3 columns
genre_split = df['genre'].str.split(',', expand=True, n=2)
genre_split.columns = ['Genre1', 'Genre2', 'Genre3']


# Keep only 1 director
df['Director1'] = df['director'].str.split(',').str[0]


# Split the 'star' column into up to 4 columns (handling newlines and commas)
star_split = df['star'].str.replace('\n', '', regex=True).str.split(',', expand=True, n=3)
star_split.columns = ['Star1', 'Star2', 'Star3', 'Star4']


# Combine the original DataFrame with the new split columns
df_combined = pd.concat([df, genre_split, star_split], axis=1)


# Drop the original 'genre', 'director', and 'star' columns if no longer needed
df_combined.drop(columns=['genre', 'director', 'star'], inplace=True)


# Save the new DataFrame to a CSV file
# df_combined.to_csv("split_columns_limited_output.csv", index=False)
```
